Text Generation

```python
from transformers import pipeline
generator = pipeline("text2text-generation", model="google/flan-t5-base")
print("Zero Shot")
print(generator("Write a story about a robot who learns emotions."))
print("few Shot")
prompt= """ Write a story
Example: A boy discovers a portal in his closet.
Example: A girl finds a talking dog in the forest.
Now: a robot who learns emotions."""
print(generator(prompt))
```

Summarization

```python
from transformers import pipeline
generator = pipeline("text2text-generation", model="google/flan-t5-base")
text = text = ("Hugging Face is a company that provides tools and models for Natural Language Processing. "
        "It created the Transformers library, which makes it easy to use state-of-the-art models.")
print("Zero Shot")
print(generator(text))
print("Chain of thought")
text2=("Summarize the following step-by-step"
       "1.Read the input"
       "2. Identify key information."
        "3. Rephrase it briefly." + text)
print(generator(text2))
```

Question & Answering

```python
from transformers import pipeline
generator = pipeline("text2text-generation", model="google/flan-t5-base")

context = ("The sun is a star at the center of the solar system. It provides light and heat to Earth.")
question = "Where is the sun located?"

print("Zero Shot")
print(generator(f"Answer the question: {question} Context: {context}"))

print("Chain of thought")
cot_qa = ("Step-by-step: Read the context. Identify key details. Then answer clearly."
          f"Context: {context} Question: {question}")
print(generator(cot_qa))
```

Translation

```python
from transformers import pipeline
generator = pipeline("text2text-generation", model="google/flan-t5-base")

print("Zero Shot")
print(generator("Translate to French: The weather is nice today."))

few_shot = ("Translate to French:"
            "English: I love cats. French: J'aime les chats."
            "English: The weather is nice today.French:")
print("Few Shot")
print(generator(few_shot))
```
